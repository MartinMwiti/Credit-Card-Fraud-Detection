# Importing Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, confusion_matrix, roc_curve
```

# Loading of the dataset

```python
data =pd.read_csv("D:\Machine Learning (Pycharm IDE)\Kaggle Competitions\Credit Card Fraud Detection\creditcard.csv")
data.drop(columns ='Time', inplace = True)
```

# Data Processing and feature selection

```python
data.info()
```

```python
data.isna().sum()
```

```python
#finding tottal number of duplicated rows
data.duplicated().sum()
```

```python
#changing dtype for the variable 'Class'
data.Class.astype('object', inplace=True)
data.describe()
```

```python
#Allow viewing of all 30 columns
pd.set_option('display.max_columns', 30)

#view first 5 rows of the dataset
data.head()
```

Finding total count for Fraud and Non fraud class then plotting them using piechart

```python
count_classes = data.Class.value_counts(sort = True).sort_index()
count_classes
#class = {0: 'Non Fraud', 1: 'Fraud'}
colors = ['g']
labels = ['Non Fraud', 'Fraud']
explode = (0.1, 0)
plt.figure(figsize= (2,7), dpi =300)
plt.pie(count_classes, explode= explode, labels= labels, colors=colors, autopct='%1.1f%%', startangle=140 )
plt.title("Fraud class piechart", loc = 'right')
```

```python
#finding correlation values for all variables in relation to variable 'Class'
Class_correl = data.corr().Class.sort_values(ascending = False)
Class_correl
```

```python
#plotting the correlations
plt.figure(figsize= (10,15), dpi =300)
fig = Class_correl.plot(kind = 'bar', legend=True)
fig.set_title('A graph of Class variable correlation with respective features')
fig.set_xlabel('Features')
fig.set_ylabel('Correlation Scale')
```

```python
#feature selection for X and y variables
X = data.drop(columns= 'Class')
y = data.Class

#splitting of the variables X and y for testing and training purposes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=42)
```

```python
#checking for variable dimensions
X_train.shape
y_train.shape
```

# Model Creation Using K-nearest neighbors

```python
knn = KNeighborsClassifier()

scores = cross_val_score(knn, X_train, y_train, cv=10)
#print('Accuracy: %0.2f%% (+/- %0.2f)' %(scores.mean()*100, (scores.std()*100)*2))
```

# Model Prediction

```python
y_pred = knn.fit(X_train, y_train).predict(X_test)
```

# Model Validation

```python
#Due to data imbalance, i will only use ROC AUC score
print('ROC AUC score: %0.2f%%' %(roc_auc_score(y_test, y_pred)*100))
```

Computing Receiver Operating Characteristic(ROC) Curve from prediction scores.

```python
#roc curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
ax= plt.subplot()
ax.plot([0, 1], [0, 1], linestyle='--')
ax.plot(fpr, tpr, marker='.')
#labeling
ax.set_xlabel('False Positive Rate or (1 - Specifity)');ax.set_ylabel('True Positive Rate or (Sensitivity)'); 
ax.set_title('Receiver Operating Characteristic(ROC) Curve') 
```

```python
cm = confusion_matrix(y_test, y_pred)
#print('confusion accuracy: %0.2f%%' %((cm.trace()/cm.sum())*100))

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax, fmt = ',', linecolor= 'lightseagreen', linewidths=3, cmap="magma"); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['Non Fraud', ' Fraud']); ax.yaxis.set_ticklabels(['Non Fraud', 'Fraud']);
```
